```python
import matplotlib.pyplot as plt
import numpy as np 
data_path = '../data/3D_spatial_network.txt'
def read_data(path):
    f = open(path,'r')
    data = []
    index = []
    for line in f:
        line = line.rstrip('\n')
        s = line.split(',')
        pt = [np.float(s[1]),np.float(s[2]),np.float(s[3])] 
        data.append(pt)
        index.append(np.int(s[0]))
    f.close()
    return np.array(index,dtype=np.int32),np.array(data)

def get_road(i,index,pt):
    return pt[index == i,:]
```

```python
index,pt = read_data(data_path)
road_id_set = set(index)
```

```python
a = get_road(index[0],index,pt)
plt.plot(a[:,0],a[:,1])
plt.show()
```

```python

for k,id in enumerate(road_id_set):
    tpt = get_road(id,index,pt)
    plt.plot(tpt[:,0],tpt[:,1])
    if k > 10000:
        break
plt.show()
```
